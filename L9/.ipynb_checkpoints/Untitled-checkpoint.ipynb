```python
import matplotlib.pyplot as plt
import numpy as np

# It may be helpful to complete this function
def norm(x, p):
    """Computes a p-norm.

    Args:
    
        x (ndarray): input array
        p (int or float): order of the norm

    Returns:
        (float): the p-norm of the array
    """
    l=0
    for i in x:
         l+=abs(i)**(p)
    return l**(1/p)

ps = [1, 2, 5, 0.5]
phi = np.linspace(0, 2 * np.pi, 500)
unit_circle = np.array([np.cos(phi), np.sin(phi)])
```

```python
for p in ps:
    plt.figure()
    u=unit_circle.copy()
    for i in range(500):
        x=np.array([u[0][i],u[1][i]])
        s=norm(x, p)
        u[0][i]=u[0][i]/s
        u[1][i]=u[1][i]/s
    plt.scatter(u[0],u[1])
    fig_1 = plt.gca()
```

```python
plt.figure()
u=unit_circle.copy()
for i in range(500):
    x=np.array([u[0][i],u[1][i]])
    s=norm(x, 1)
    u[0][i]=u[0][i]/s
    u[1][i]=u[1][i]/s
plt.scatter(u[0],u[1])
fig_1 = plt.gca()
```
